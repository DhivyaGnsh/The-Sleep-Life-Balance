```python
import pandas as pd
```

```python
data = {
    'Species': ['A', 'B', 'A', 'C', 'B'],
    'Growth_Rate': [1.2, 0.8, 1.5, 1.1, 0.9],
    'Temperature': [20, 25, 22, 18, 24],
    'Sunlight': [8, 9, 10, 7, 8],
    'Biomass': [100, 80, 120, 90, 85]
}
```

```python
df = pd.DataFrame(data)
```

```python
avg_growth_rate = df['Growth_Rate'].mean()
print("Average Growth Rate:", avg_growth_rate)
```

To find the correlation between growth rate and sunlight

```python
correlation = df['Growth_Rate'].corr(df['Sunlight'])
```

```python
print("Correlation between Growth and Sunlight:", correlation)
```

Average sleep duration for each BMI category

```python
data = pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
```

```python
df = pd.DataFrame(data)
```

```python
avg_sleep_by_bmi = data.groupby("BMI Category")["Sleep Duration"].mean()
print(avg_sleep_by_bmi)
```

Percentage of male and female in data

```python
gender_distribution = data["Gender"].value_counts(normalize=True) * 100
print(gender_distribution)
```

correlation between daily steps and heart rate

```python
correlation = data["Daily Steps"].corr(data["Heart Rate"])
print(f"Correlation between daily steps and heart rate: {correlation}")
```
